```python
# importing libraries
import pandas as pd
import numpy as np
from sodapy import Socrata
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
import seaborn as sns
import urllib.parse
import json
from urllib.request import urlopen
import requests # library to handle requests
from datetime import datetime

# Displays up to 30 columns
pd.set_option('display.max_columns', 30)
pd.set_option('display.max_rows', 50)
```

```python
# Data snippet; Note the data info will vary
data='bqiq-cu78'

# query parameter as a string
query = """
    SELECT
        *
    WHERE
        bias_motive_description IN ('ANTI-JEWISH','ANTI-ASIAN')
        AND complaint_year_number BETWEEN '2019' AND '2021'
        AND month_number=9
        AND arrest_id IS NOT NULL
    ORDER BY
        county DESC
    LIMIT
        2000
    """

url=f'https://data.cityofnewyork.us/resource/{data}.json?$query={query}'
df= requests.get(url).json()
df=pd.DataFrame(df)
print(df.shape)
df
```

```python
#GROUP BY
# Data snippet; Note the data info will vary
data='bqiq-cu78'

# query parameter as a string
query = """
    SELECT
        county,
        count(*) as Count
    WHERE
        bias_motive_description IN ('ANTI-JEWISH','ANTI-ASIAN')
    GROUP BY
        county
    HAVING
        Count >100
    ORDER BY
        county DESC
    LIMIT
        2000
    """

url=f'https://data.cityofnewyork.us/resource/{data}.json?$query={query}'
df= requests.get(url).json()
df=pd.DataFrame(df)
print(df.shape)
df
```
